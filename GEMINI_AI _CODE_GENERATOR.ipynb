```python
# 1. Install libraries
!pip install google-genai ipywidgets -q

import os
from google import genai
from google.genai import types
from google.colab import userdata
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output

# 2. Authenticate
try:
    api_key = userdata.get('GEMINI_API_KEY')
    client = genai.Client(api_key=api_key)
except Exception as e:
    print("Error: Make sure you added 'GEMINI_API_KEY' to your Colab Secrets side panel!")

# 3. Build UI Layout
prompt_input = widgets.Textarea(
    value='',
    placeholder='Type your code request here (e.g., "Write a Python script to sort a dictionary")...',
    description='Prompt:',
    layout=widgets.Layout(width='90%', height='120px')
)

generate_btn = widgets.Button(
    description='Generate Code',
    button_style='success',
    tooltip='Click to run AI',
    icon='code'
)

output_area = widgets.Output()

# 4. Core Bot Logic
def on_button_clicked(b):
    with output_area:
        clear_output()
        user_prompt = prompt_input.value.strip()

        if not user_prompt:
            print("Please type an instruction first!")
            return

        print("🤖 Processing on Google Cloud... Generating code...")

        try:
            response = client.models.generate_content(
                model='gemini-2.5-flash',
                contents=user_prompt,
                config=types.GenerateContentConfig(
                    system_instruction=(
                        "You are an expert AI software engineer. Provide high-quality, optimal, "
                        "and well-commented code. Always wrap the code blocks cleanly in markdown."
                    ),
                    temperature=0.2,
                ),
            )
            clear_output()
            display(Markdown(response.text))

        except Exception as e:
            clear_output()
            print(f"Error: {e}\nDouble-check your API key secret toggle is turned ON.")

# 5. Display App
generate_btn.on_click(on_button_clicked)
display(widgets.VBox([prompt_input, generate_btn, output_area]))
```
